```python
def cm(x,y):
    TP=0
    TN=0
    FN=0
    FP=0
    
    mat1 = x
    mat2 = y
    for i in range(len(mat1)): 
        if mat1[i]==mat2[i]==1:
            TP += 1
        if mat1[i]==mat2[i]==0:
            TN += 1
        if mat1[i]==1 and mat2[i]==0:
            FN += 1
        if mat1[i]==0 and mat2[i]==1:
            FP += 1

    print(TP)
    print(TN)
    print(FN)
    print(FP)
    print("total:", len(mat1))
    print(TP/len(mat1)*100)
 
```
